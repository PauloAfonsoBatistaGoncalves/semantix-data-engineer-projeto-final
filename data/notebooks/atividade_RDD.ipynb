```python
palavras = rdd.flatMap(lambda x: x.split(" "))
palavras.collect()
```

```python
palavras
```

```python
palavras_minusculas = palavras.map(lambda palavra: palavra.lower())
palavras_minusculas.collect()
```

```python
palavras_maiores_que_2 = palavras_minusculas.filter(lambda palavra: len(palavra) >= 2)
palavras_maiores_que_2.collect()
```

```python
p_chave_valor = palavras_maiores_que_2.map(lambda palavra : (palavra,1))
p_chave_valor.collect()
```

```python
p_reduce= p_chave_valor.reduceByKey(lambda key1, key2: key1 + key2)
p_reduce.collect()
```

```python
p_ordena = p_reduce.sortBy(lambda palavra: palavra[0], True)
p_ordena.collect()
```

```python
p_ordena2 = p_reduce.sortBy(lambda palavra: palavra[1], False)
p_ordena2.collect()
```

```python
p_apenas_1 = p_ordena2.filter(lambda p: p[1] > 1)
p_apenas_1.collect()
```
